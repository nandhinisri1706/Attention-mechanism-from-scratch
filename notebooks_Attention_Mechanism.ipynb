```python
import numpy as np 
x = np.array([
    [1.0, 0.0, 1.0],
    [0.0, 2.0, 1.0],
    [1.0, 1.0, 0.0]
])
print("Input embeddings (x):")
print(x)

# step 2 
np.random.seed(42)
d_model = x.shape[1]
d_k = 3
WQ = np.random.randn(d_model, d_k)
WK = np.random.randn(d_model, d_k)
WV = np.random.randn(d_model, d_k)

#Step 3
Q = x @ WQ 
K = x @ WK 
V = x @ WV
print("\nQuery(Q) :")
print(Q)
print("\nKey (K) : ")
print(K)
print("\nvalue (V) : ")
print(V)

#step 4 
Scores = Q @ K.T
print("\nAttention Scores : " )
print(Scores)

#step 5
Scalled_Scores = Scores / np.sqrt(d_k)
print("\nScalled Scores : ")
print(Scalled_Scores)

#step 6 
def softmax(x) :
    exp_x = np.exp(x- np.max(x, axis = -1, keepdims = True))
    return exp_x / np.sum(exp_x, axis = -1, keepdims = True)

attention_weights = softmax(Scalled_Scores)
print("\nAttention Weights :")
print(attention_weights)

#Step 7 
attention_output = attention_weights @ V
print("\nFinal Attention output : ")
print(attention_output)
```
